```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# to read the excel file from usiong csv read
data=pd.read_csv("california.csv")
data.head()
```

```python
#to make a histogram ploting using hist function.
data["households"].hist()
plt.show()
```

```python
# To make a line chart using line function
data.plot.line(x='population',y='median_income',figsize=(8,6))
plt.show()
```

```python
#Ploting graph using scatter in python pandas
data.plot.scatter(x='population',y='median_income',figsize=(8,6))
plt.show()
```

```python
# using boxplt in python pandas
data.plot.box(figsize=(8,6))
plt.show()
```

```python
# Pour faire hexagonal ploting using pandas
data.plot.hexbin(x='population',y='median_income',figsize = (6,5))
plt.show()
```

```python
#To plot a pie chart in pandas
dataf=pd.DataFrame({'cost':[79,40,60]},index=['Oranges','Bananas','Apple'])
dataf.plot.pie(y='cost',figsize=(8,6))
plt.show()
```
